```python
from collections import deque
queue = deque()

def enqueue(data):
    queue.append(data)

def dequeue():
    if queue:
        data = queue.popleft()
    else:
        data = -1
    return data

def getFront():
    if queue:
        return queue[0]
    else:
        return -1

print("=============================================================")
print(" 1. Enqueue      2. Dequeue      3. Get Front       0. Exit ")
print("=============================================================")

print('시작')

while True:
    # 메뉴 프린트
    # 메뉴 번호 입력
    menu = int(input('>>'))
               
    if menu == 0:break
    elif menu == 1: # enqueue
        data = int(input('추가할 데이터 :'))
        enqueue(data)
        
    elif menu == 2: # dequeue
        data = dequeue()
        if data != -1:
           print('가져온 데이터 :', data)
        else:
           print('정수형 Queue 종료')
            
    elif menu == 3: # peek
         data = getFront()
         if data != -1:
             print(data)
         else:
             print('정수형 Queue 종료')

    print('[Queue]', queue)

print('종료')
```
