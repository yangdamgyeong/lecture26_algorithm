### 소수

```python
# 소수를 함수로
def is_prime_number(N):
    if N <= 1:
        return False   #if N == 1: return False 도 가능
        
    for i in range(2, N):
        if N % i == 0:
            return False
    
    return True
print(is_prime_number(0)) 
print(is_prime_number(9)) 
```

```python
def is_prime_number(N):
    if N == 1: return False

    is_prime = True
    for i in range(2, N):
        if N % 1 == 0:
            is_prime = False
            break
    return is_prime
```

```python
# 알고리즘 개선하기 (1)
def is_prime_number(N):
    if N == 2: return True:
    if N == 1 or N % 2 == 0: return False

    for i in range(3, N, 2):
        if N % i == 0:
            return False
    return True
```

```python
# 정수 N을 받아서 base에 대한 진법의 수로 변환
def convert_base(N, base):
    # N을 base로 나눈 몫을 몫이 0이 아닌 동안 계속 N으로 나눈다.
    # 나눈 나머지를 진법에 해당하는 숫자를 변환하여 리스트에 저장한다.
    # N : 몫을 저장
    # new_list :. 나머지를 저장하는 리스트
    new_list = []
    digits = "0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    while N != 0:
        new_list.append(digits[N % base])
        N //= base
    return new_list

print(convert_base(25,2))
```

```python
def convert_base(N, base):
    new_number = ''
    digits = "0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    while N != 0:
        new_number = digits[N % base] + new_number
        N //= base
    return new_number
print(convert_base(25,2))
```
