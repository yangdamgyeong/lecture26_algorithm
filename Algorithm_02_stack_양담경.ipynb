```python
# 데이터 정의
stack = []

while True:
    # 메뉴를 출력
    print('''
    =======================================
    1.Push 2.Pop 3.Peek 0.Exit
    =======================================
    ''')
    # 입력을 받는다
    menu = int(input(' >> 연산 메뉴 선택 :'))
    # print(menu)
    
    # 입력에 따라 stack 연산을 수행하고
    # 스택의 상태를 출력한다
    if menu == 0:         
        break
    elif menu ==1: #push
        data = int(input('[Push] 추가할 데이터 입력 :'))
        stack.append(data)
    elif menu == 2: #pop
        data = stack.pop()
        print('[Pop] 가져온 데이터:', data)
    elif menu == 3: #peek
         data = stack[len(stack)-1]
         print('top 데이터 확인 :', data)
         continue

print('[Stack]', stack)
    
print('정수형 스택 종료')    
# 위의 과정을 0. Exit을 선택할 때까지 반복
```
