```python
# selection sort
# 정렬이 안된 부분 (start를 포함한 그 이후)에서 가장 작은 수를 찾아서 start에 있는 수와 swap
# start += 1 하나씩 증가
def select(A, start):
    min = A[start]
    min_index = start

    for i in range(start+1, len(A)):
        if A[i] < min:
            min = A[i]
            min_index = i

    return min_index
    
def selection_sort(data_list):
    for start in range(len(data_list)-1):
        min_index = select(data_list, start)
        data_list[start], data_list[min_index]=data_list[min_index], data_list[start]

A = [23, 5, 44, 1, 6]
selection_sort(A)
print(A)
```

```python
def print_step(step, A):
    print(f'{step:2} 단계 : {A}')
    
def select(A, start):
    min = A[start]
    min_index = start

    for i in range(start+1, len(A)):
        if A[i] < min:
            min = A[i]
            min_index = i

    return min_index
    
def selection_sort(data_list):
    for start in range(len(data_list)-1):
        min_index = select(data_list, start)
        data_list[start], data_list[min_index]=data_list[min_index], data_list[start]
        print_step(start, data_list)

A = [23, 5, 44, 1, 6]
print('원래 데이터 : ', A)
selection_sort(A)
print('정렬후 데이터 :', A)
```

```python
# insertion Sort
def insertion_sort(A):
    for cur_index in range(1, len(A)):
        # 해당 데이터의 위치를 찾아서 insertion
        key = A[cur_index]
        i = cur_index - 1 # 내 데이터보다 왼쪽에서 시작할거야
        while i >= 0 and A[i] > key: # 내 데이터가 0 보다 크거나 같을 때
            A[i+1] = A[i]  # 찾아가면서...
            i -= 1
        # 작거나 같은 데이터의 오른쪽에 저장
        A[i+1] = key
        # print_step(cur_index, A)
```

```python
# bubble Sort
def bubble_sort(A):
    for last in range(len(A)-1, 0, -1):
        is_changed = False
        for i in range(last):
            if A[i] > A[i+1]:
                A[i], A[i+1] = A[i+1], A[i]
                is_changed = True
        print_step(last, A)
        if not is_changed: return
```

```python
A = [23, 5, 44, 1, 6]
print('원래 데이터 : ', A)
#selection_sort(A)
#insertion_sort(A)
bubble_sort(A)
print('정렬후 데이터 :', A)
```
