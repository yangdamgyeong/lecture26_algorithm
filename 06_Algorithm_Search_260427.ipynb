### 이진 탐색(Binary Search)

```python
#recursive call를 활용한 binary search
def binary_search_recursive(A, key, low, high):
    if (low > high): #그만 할 조건
        return -1  # 찾는 key가 A에 없는 경우=탐색할 항목이 없음

    # 중앙값을 찾는다
    middle = (low+high) // 2

    # 중앙에 있는 값과 비교하여 같으면 해당 index를 return
    if key == A[middle]:
        return middle
    elif key < A[middle]:  #탐색 영역을 왼쪽으로 설정
        return binary_search_recursive(A, key, low, middle-1)
    else: # 왼쪽도 아니고 같지도 않으면 = 즉 오른쪽이면 
        return binary_search_recursive(A, key, middle+1, high)

data = [2, 6, 11, 13, 18, 20, 22, 27, 29, 30, 34, 38, 41, 42, 45, 47]
num = int(input('찾을 숫자 :'))
index = binary_search_recursive(data, num, 0, len(data))
if index == -1:
    print('없는 숫자입니다.')
else:
    print(f'>>> 숫자 {num}은 index {index}에 있습니다.')        
```

```python
# while loop를 활용한 binary search
def binary_search(A, key):
    low = 0 #low는 인덱스의 첫번째 위치= 탐색 영역의 시작
    high = len(A) - 1 # 탐색영역의 끝
    while low <= high: # 영역을 바꿔가면서 위치 찾을거임
        # 중앙값을 찾는다
        middle = (low + high) // 2
        # 중앙에 있는 값과 비교하여 같으면 해당 index를 return
        if A[middle] == key:
            return middle
        # 작으면 왼쪽 영역 탐색 high 값 변경
        elif A[middle] > key:
            high = middle - 1
        # 크면 오른쪽 영역 탐색 low 값 변경
        else:
            low = middle + 1
    return -1 # 찾는 데이터 없음
data = [2, 6, 11, 13, 18, 20, 22, 27, 29, 30, 34, 38, 41, 42, 45, 47]
num = int(input('찾을 숫자 :'))
index = binary_search(data, num)
if index == -1:
    print('없는 숫자입니다.')
else:
    print(f'>>> 숫자 {num}은 index {index}에 있습니다.')
```
